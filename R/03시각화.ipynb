## 데이터 시각화
+ 데이터를 관찰하는 과정에서 선택사항이 아니라 반드시 거쳐야 하는 필수 과정

+ 그래프와 그림형태로 제시하면 데이터의 관계구분 용이, 열정적인 반응 동반

## R 기본 그래프 함수
+ plot, hist, boxplot, barplot, pie, stem

```r
# 표준 정규분포를 따르는 난수 15개 생성하고, 그에따른 그래프를 작성해서 시각화
# 정규분포 : 수집한 자료가 평균0, 표준편차가 1인 형태
# rnorm(갯수) : -1~1 사이의 난수 생성 함수

x <- rnorm(15, mean=0, sd=1)
x
summary(x)
```

```r
sd(x)
```

```r
# 산점도 : plot(x, y, 옵션)
plot(x)        # 변수가 1개일때는 자동으로 y축에 할당
```

```r
# 막대그래프 : barplot(x, y, 옵션)
barplot(x)
```

```r
# 히스토그램 : hist(x, 구간, 옵션)
hist(x)
```

```r
# 상자수염 그래프 : boxplot(x, 옵션)
boxplot(x)
```

```r
# 줄기잎 그래프 : stem(x, 옵션)
stem(x)
```

## 그래프 다양한 설정 보기

+ xlim/ylim : 축값 범위
+ col : 그래프 색상
+ type : 그래프 유형

## 그래프 눈금 세부 조정하기
+ 그래프 그릴때 axes 속성을 FALSE로 지정
+ axis 함수로 세부 조정
+ axis(축지정, 눈금지정)

```r
plot(heights, main='student heights', sub='class A', xlab='numbers', col='blue', pch=3, axes=F)

axis(side=1, at=1:10)  # x축 설정
axis(side=2, at=seq(140,185,5), las=2)  # y축 설정
box()  # 그래프 테두리 표시
grid()
```

## 그래프 텍스트 추가하기
+ 특정 x,y위치에 글자 표시
+ text(y좌표, 글자크기, 글자위치, 글자내용)
+ adj : 음수 - 오른쪽위, 양수 - 왼쪽아래

```r
plot(heights, col='red')

text(heights, cex=0.7, adj=c(0, -0.2), lab=heights)
grid()
```

```r
# jupyter R plot에 한글 사용하기
library(sysfonts)

font_add('NanumGothic', '/usr/share/fonts/NanumFont/NanumGothicBold.ttf')
par(family='NanumGothic')
```

```r
# A 약물에 따른 환자 반응 - 눈금조정, 텍스트 추가
plot(dose, drugA, main='약물에 따른 환자 반응 A', sub='나이', xlab='age', ylab='effect', col='red', axes=F)


axis(side=1, at=seq(20,60,5))  # x축 설정
axis(side=2, at=seq(15,60,5), las=2)

text(dose, drugA, cex=1, adj=c(0, -0.2), lab=drugA)
box()
grid()
```

## 산점도 행렬
+ 여러 변수 쌍의 관계를 동시에 평가할 수 있음
+ plot(데이터프레임명)

```r
head(iris)

plot(iris)  # 범주형인 species가 포함됨

plot(iris[,-5])   # 범주형인 species는 제외
```

```r
# iris 데이터셋에서 petal Length/Width로 시각화
plot(iris$Petal.Length, iris$Petal.Width)
```

```r
# iris 데이터셋에서 petal Length/Width로 시각화
# 독립변수에 따른 종속변수의 변화도 시각화
str(iris)
```

```r
# 범주형 데이터를 토대로 붓꽃의 품종을 ○□+ 기호로 표현
plot(iris$Petal.Length, iris$Petal.Width, pch=as.integer(iris$Species))
```

```r
# 범주형 데이터를 토대로 붓꽃의 품종을  색상으로 표현
# 1.black, 2:red, 3:greed, 4:blue
plot(iris$Petal.Length, iris$Petal.Width, col=as.integer(iris$Species)+1)     
```

```r
# 범주형 데이터를 토대로 붓꽃의 품종을 색상과 기호로 나타낼수 있음
plot(iris$Petal.Length, iris$Petal.Width, 
     pch=as.integer(iris$Species), 
     col=as.integer(iris$Species)+1)     
```

# 범례 표시하기
+ legend(위치, 값, 기호, 색상)
+ 위치: top, bottom, right, left, center

```r
plot(iris$Petal.Length, iris$Petal.Width, 
     pch=as.integer(iris$Species), 
     col=as.integer(iris$Species)+1) 
legend('topleft', c('setosa', 'verdicolor', 'verginica'), pch=c('○','△','+'), col=c('red','green','blue'))
```

```r
# 자동으로 텍스트, 기호, 색상 지정

plot(iris$Petal.Length, iris$Petal.Width, 
     pch=as.integer(iris$Species), 
     col=as.integer(iris$Species)+1) 

legend('bottomright', levels(iris$Species), pch=unique(as.integer(iris$Species)), col=unique(as.integer(iris$Species)+1))
```

## 2개의 산점도를 동시에 출력
+ plot 함수와 lines 함수를 번갈아가며 사용
+ 이런경우 범례를 반드시 출력할 것!

```r
# A,B 약물에 따른 환자 반응
plot(dose, drugA, col='red', type='b', pch=1, ylab='drugs')
lines(dose, drugB, col='blue', type='b', pch=2)
legend('topleft', c('dragA', 'drugB'),
        pch=c(1,2), col=c(2,4))
```

## 매니저별 성별/국적별 나이에 대한 산점도 작성

```r
manager <- c(1,2,3,4,5)
country <- c('US','US','UK','UK','UK')
gender <- c('M','F','F','M','F')
age <- c(32, 45, 25, 39, 99)
```

## 벡터들로 데이터 프레임 생성하기
+ data.frame(벡터1, 벡터2, 벡터3, ...)

```r
manager <- data.frame(manager, country, gender, age)
manager
```

```r
# 국적별 나이

plot(manager$age, pch=as.integer(manager$country), 
     col=as.integer(manager$country)+1)

legend('topleft', levels(manager$country), pch=unique(as.integer(manager$country)), col=c('red','green'))
```

```r
# 성별 나이

plot(manager$age, pch=as.integer(manager$gender), 
     col=as.integer(manager$gender)+1)

legend('topleft', levels(manager$gender), pch=unique(as.integer(manager$gender)), col=c('red', 'green'))
```

```r
# 국적별 나이
plot(manager$age, ylab='나이', pch=as.integer(manager$country), col=as.integer(manager$country)+1)

legend('topleft', levels(manager$country), pch=c('○','△','+','x','◇'), col=c('red','green','blue','cyan','magenta'))
```
